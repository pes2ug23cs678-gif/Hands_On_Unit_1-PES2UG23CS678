```python
from transformers import pipeline, set_seed
import os
import nltk
import warnings
```

```python
set_seed(100)
```

```python
def smalltrain(model_name):
  warnings.filterwarnings('ignore')
  if model_name in ['bert-base-uncased', 'bert-base-cased']:
    sentence = "The article is about [MASK] and AI."
  else:
    sentence = "The article is about <mask> and AI."
  model = pipeline("fill-mask", model=model_name)
  result = model(sentence)
  for i in range(len(result)):
    print(f"{result[i]['sequence']}")
```

```python
smalltrain('roberta-base')
```

```python
smalltrain('bert-base-cased')
```

```python
smalltrain('facebook/bart-base')
```
